```python
info = ""
n = int(input("Enter n = "))
for i in range(1,n+1):
    name = input("Enter name = ")
    age = int(input("Enter age = "))
    add = input("Enter add = ")
    phone = input("Enter phone = ")
    info = info + f"{i},{name},{age},{add},{phone}\n"

file = open("info.csv","w")
file.write("sn,name,age,add,phone\n")
file.write(info)
file.close()
```
